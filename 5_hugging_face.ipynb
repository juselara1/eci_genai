## Entrenamiento de Modelo

### Conjunto de Datos

```python
ds = datasets.load_dataset("bertin-project/alpaca-spanish")
ds
```

```python
ds_full = ds["train"]
```

```python
ds_full[:2]
```

```python
list(map(lambda x: x ** 2, range(10)))
```

```python
ds_template = ds_full.map(
    lambda sample: {
        "text": f"""[PREGUNTA]
{sample['instruction']} {sample["input"]}
[RESPUESTA]
{sample['output']}"""
    },
    remove_columns=ds_full.column_names,
    num_proc=4
)
```

```python
ds_template
```

```python
print(ds_template[0]["text"])
```

```python
print(ds_template[1]["text"])
```

```python
ds_partition = ds_template.train_test_split(test_size=0.3, seed=0)
ds_partition
```

### Tokenizador

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
def tokenize(sample: dict) -> dict:
    inputs = tokenizer(sample["text"], truncation=True, max_length=128)
    return {"input_ids": inputs["input_ids"]}
```

```python
ds_token = ds_partition.map(
    tokenize, remove_columns=ds_template.column_names, num_proc=4
).with_format("torch")
```

```python
ds_token["train"][0]
```

### Modelo

```python
config = GPT2Config.from_pretrained("gpt2", n_ctx=128)
```

```python
model = GPT2LMHeadModel(config).to("cuda")
model
```

```python
inspect.getmro(GPT2LMHeadModel)
```

```python
sum(p.numel() for p in model.parameters())
```

### Entrenamiento

```python
tokenizer.eos_token
```

```python
tokenizer.pad_token = tokenizer.eos_token
```

```python
data_collator = DataCollatorForLanguageModeling(tokenizer, mlm=False)
```

```python
inputs = data_collator(ds_token["train"][:3]["input_ids"])
```

```python
print(tokenizer.batch_decode(inputs["input_ids"][2:3])[0])
```

```python
print(tokenizer.batch_decode(inputs["labels"][2:3])[0])
```

```python
inputs.keys()
```

```python
args = TrainingArguments(
    output_dir="gpt2_custom",
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    num_train_epochs=1,
    warmup_steps=1_000,
    fp16=True,
    learning_rate=1e-4,
    save_strategy="epoch",
    report_to="none"
)
```

```python
trainer = Trainer(
    model=model,
    processing_class=tokenizer,
    args=args,
    data_collator=data_collator,
    train_dataset=ds_token["train"].select(list(range(1000))),
    eval_dataset=ds_token["test"].select(list(range(1000))),
)
```

```python
trainer.train()
```
